```python
import imageio
import os
from PIL import Image

imageList = [x for x in os.listdir('../data/good') if x.endswith('.png')]
```

```python
def is_white(pixel):
    for x in pixel:
        if x != 255: 
            return False
    return True

print(is_white([255, 255, 255]))
print(is_white([255, 255, 255, 255]))
print(is_white([255, 13, 255, 255]))
```

```python
def crop_image(fileName, infolder, outfolder):
    print(fileName)
    im = imageio.imread(infolder + fileName)

    #display(Image.fromarray(im))
    print(im.shape)

    middleH = im.shape[0] // 2
    middleT = im.shape[1] // 2
    height = int(im.shape[1])
    size = 622, int(im.shape[0])

    row_data = im[middleH]

    start = None
    stop = None

    for col in range(32, len(row_data)):        
        if start == None:
            if not is_white(row_data[col]):
                start = col

        elif stop == None:
            if is_white(row_data[col]):
                if (col < start + 50):
                    start = None
                    stop = None
                else:
                    stop = col
                    break

    startV = None
    stopV = None

    for col in range(im.shape[0]):
        pixel = list(im[col][middleT])

        if startV == None:
            if not is_white(pixel):
                startV = col
        
        elif stopV == None:
            if is_white(pixel):
                stopV = col
                break

    pil_image = Image.fromarray(im)

    outputImg = pil_image.crop((start+1, startV+1, stop-1, stopV-1))
    outputImg = outputImg.resize(size, Image.ANTIALIAS)
    outputImg.save(outfolder + fileName)
    
crop_image('waterfall_1186_2017-02-03T03-02-05.png', '../data/good/', '../data/cropped/good/')
```

```python
images_good = [x for x in os.listdir('../data/good') if x.endswith('.png')]
for img in images_good:
    crop_image(img, '../data/good/', '../data/cropped/good/')
```

```python
images_bad = [x for x in os.listdir('../data/bad') if x.endswith('.png')]
for img in images_bad:
    crop_image(img, '../data/bad/', '../data/cropped/bad/')
```
